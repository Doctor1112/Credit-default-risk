```python
def cat_agg(df, group):
    group_var = df[group]
    df = df.select_dtypes(include='object')
    df = pd.get_dummies(df)
    df[group] = group_var
    df_group = df.groupby(group).agg(['sum', 'mean']).reset_index()
    
    columns = [group]
    for i in df_group.columns.levels[0][:-1]:
        for j in df_group.columns.levels[1][:-1]:
            columns.append(i + '_' + j)
    df_group.columns = columns
    return df_group
```

```python
def num_agg(df, group):
    group_var = df[group]
    df = df.select_dtypes(exclude='object')
    df[group] = group_var
    df_group = df.groupby(group).agg(['sum', 'mean', 'min', 'max', 'count']).reset_index()
    columns = [group]
    for i in df_group.columns.levels[0][:-1]:
        for j in df_group.columns.levels[1][:-1]:
            columns.append(i + '_' + j)
    df_group.columns = columns
    df_group['count'] = df.groupby(group).count().iloc[:, 0].values
    return df_group
    
```
